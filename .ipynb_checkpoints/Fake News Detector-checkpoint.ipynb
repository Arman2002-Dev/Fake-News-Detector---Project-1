```python
"""
PREPARING DATA
1. Combine title and text for better analysis
2. Convert FAKE/REAL to 1/0
"""

# Create a copy
df_processed = df.copy()

# Combine title and text (title gives important context!)
df_processed['combined_text'] = df['title'].fillna('') + ' ' + df['text'].fillna('')

# Convert labels: FAKE=1, REAL=0
df_processed['label_num'] = df['label'].apply(lambda x: 1 if x == 'FAKE' else 0)

print("✅ Data prepared!")
print(f"   Total samples: {len(df_processed)}")
print(f"   FAKE (1): {(df_processed['label_num'] == 1).sum()}")
print(f"   REAL (0): {(df_processed['label_num'] == 0).sum()}")

# Keep only what we need
df_final = df_processed[['combined_text', 'label_num']].copy()
df_final.columns = ['text', 'label']  # Rename for simplicity

print(f"\nSample combined text (first 200 chars):")
print(df_final['text'].iloc[0][:200] + "...")
```

```python
"""
CLEANING TEXT
"""

import re
from nltk.corpus import stopwords
import nltk

# Download if not already done
nltk.download('stopwords', quiet=True)

stop_words = set(stopwords.words('english'))

def clean_text(text):
    """
    Clean and preprocess text
    """
    # Convert to lowercase
    text = str(text).lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove special characters and digits, keep only letters and spaces
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove extra whitespace
    text = ' '.join(text.split())
    
    # Remove stopwords and short words
    words = text.split()
    words = [w for w in words if w not in stop_words and len(w) > 2]
    
    return ' '.join(words)

print("Cleaning text...")
print("⏳ This may take a minute for 6000+ articles...")

df_final['cleaned_text'] = df_final['text'].apply(clean_text)

# Remove empty rows
df_final = df_final[df_final['cleaned_text'].str.len() > 10]

print(f"✅ Cleaning complete!")
print(f"   Articles remaining: {len(df_final)}")
print(f"\nBefore cleaning:")
print(df_final['text'].iloc[0][:150] + "...")
print(f"\nAfter cleaning:")
print(df_final['cleaned_text'].iloc[0][:150] + "...")
```

```python
"""
SPLITTING DATA
80% for training, 20% for testing
"""

from sklearn.model_selection import train_test_split

X = df_final['cleaned_text']
y = df_final['label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 20% test
    random_state=42,    # Reproducible
    stratify=y          # Keep same FAKE/REAL ratio
)

print(f"✅ Data split!")
print(f"   Training: {len(X_train)} articles")
print(f"   Testing: {len(X_test)} articles")
print(f"\nTraining set - REAL: {(y_train==0).sum()}, FAKE: {(y_train==1).sum()}")
print(f"Testing set - REAL: {(y_test==0).sum()}, FAKE: {(y_test==1).sum()}")
```

```python
"""
VECTORIZATION
Convert text to numbers using TF-IDF
"""

from sklearn.feature_extraction.text import TfidfVectorizer

print("Converting text to numbers...")

tfidf = TfidfVectorizer(
    max_features=5000,      # Top 5000 words
    stop_words='english',   # Remove common words
    ngram_range=(1, 2),     # Single words + pairs
    min_df=2,               # Ignore rare words
    max_df=0.8              # Ignore very common words
)

# Fit on training data, transform both
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

print(f"✅ Vectorization complete!")
print(f"   Vocabulary size: {len(tfidf.get_feature_names_out())}")
print(f"   Training matrix: {X_train_tfidf.shape}")
print(f"   Testing matrix: {X_test_tfidf.shape}")
```

```python
"""
TRAINING LOGISTIC REGRESSION
"""

from sklearn.linear_model import LogisticRegression

print("Training model...")
print("⏳ This may take 30-60 seconds...")

model = LogisticRegression(
    max_iter=1000,
    random_state=42,
    C=1.0,
    class_weight='balanced'
)

model.fit(X_train_tfidf, y_train)

print("✅ Model trained!")

# Show most important words
feature_names = tfidf.get_feature_names_out()
coefficients = model.coef_[0]

print(f"\n🔴 Top words indicating FAKE news:")
top_fake = np.argsort(coefficients)[-10:]
for idx in reversed(top_fake):
    print(f"   {feature_names[idx]}: {coefficients[idx]:.3f}")

print(f"\n🟢 Top words indicating REAL news:")
top_real = np.argsort(coefficients)[:10]
for idx in top_real:
    print(f"   {feature_names[idx]}: {coefficients[idx]:.3f}")
```

```python
"""
EVALUATION
"""

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Predict
y_pred = model.predict(X_test_tfidf)
y_prob = model.predict_proba(X_test_tfidf)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"🎯 Accuracy: {accuracy*100:.2f}%")

# Detailed report
print(f"\n📊 Classification Report:")
print(classification_report(y_test, y_pred, target_names=['REAL', 'FAKE']))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['REAL', 'FAKE'], 
            yticklabels=['REAL', 'FAKE'])
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

# Specific metrics
tn, fp, fn, tp = cm.ravel()
print(f"\nDetailed:")
print(f"   True Negatives (correctly identified REAL): {tn}")
print(f"   True Positives (correctly identified FAKE): {tp}")
print(f"   False Positives (REAL called FAKE): {fp}")
print(f"   False Negatives (FAKE called REAL): {fn}")
```

```python
"""
CELL 14: IMPORT IMAGE LIBRARIES
--------------------------------
"""

import cv2  # Computer vision
from PIL import Image  # Image processing
import pytesseract  # OCR - Extract text from images
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import io
import base64

# Configure pytesseract path for Ubuntu
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

print("✅ Image libraries imported!")
print(f"OpenCV version: {cv2.__version__}")
print(f"Tesseract path: {pytesseract.pytesseract.tesseract_cmd}")
```

```python
"""
CELL 15: IMAGE PREPROCESSING FUNCTIONS
---------------------------------------
Functions to clean and prepare images for OCR
"""

def preprocess_image_for_ocr(image_path):
    """
    Clean image to make text extraction easier
    """
    # Read image
    img = cv2.imread(image_path)
    
    if img is None:
        raise ValueError(f"Could not load image: {image_path}")
    
    # Convert to grayscale (black and white)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Remove noise
    denoised = cv2.fastNlMeansDenoising(gray, None, 10, 7, 21)
    
    # Increase contrast
    _, binary = cv2.threshold(denoised, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    # Resize if too small (helps OCR)
    height, width = binary.shape
    if height < 1000:
        scale = 1000 / height
        binary = cv2.resize(binary, None, fx=scale, fy=scale, interpolation=cv2.INTER_CUBIC)
    
    return binary

def extract_text_from_image(image_path):
    """
    Extract text from image using OCR
    """
    try:
        # Preprocess image
        processed_img = preprocess_image_for_ocr(image_path)
        
        # Extract text with OCR
        # config: 
        #   --psm 6 = Assume a single uniform block of text
        #   -l eng = English language
        custom_config = r'--oem 3 --psm 6 -l eng'
        
        text = pytesseract.image_to_string(processed_img, config=custom_config)
        
        # Clean extracted text
        text = text.strip()
        text = ' '.join(text.split())  # Remove extra whitespace
        
        return text, processed_img
        
    except Exception as e:
        return f"Error: {str(e)}", None

def display_image_with_text(image_path, extracted_text):
    """
    Show original image and extracted text side by side
    """
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # Original image
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[0].imshow(img_rgb)
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    
    # Processed image
    processed = preprocess_image_for_ocr(image_path)
    axes[1].imshow(processed, cmap='gray')
    axes[1].set_title('Processed for OCR')
    axes[1].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n📝 Extracted Text (first 300 chars):")
    print(extracted_text[:300] + "..." if len(extracted_text) > 300 else extracted_text)
    print(f"\n📊 Total characters extracted: {len(extracted_text)}")

print("✅ Image processing functions defined!")
```

```python
"""
CELL 17: UNIFIED PREDICTION FUNCTION
------------------------------------
Works with both text input and image input
"""

def predict_from_text(title, article_text):
    """
    Predict from text input (from Cell 12)
    """
    combined = str(title) + ' ' + str(article_text)
    cleaned = clean_text(combined)
    
    if len(cleaned) < 10:
        return {
            'error': 'Text too short after cleaning',
            'prediction': None,
            'confidence': 0
        }
    
    vectorized = tfidf.transform([cleaned])
    prediction = model.predict(vectorized)[0]
    prob = model.predict_proba(vectorized)[0]
    confidence = max(prob) * 100
    
    return {
        'input_type': 'TEXT',
        'is_fake': bool(prediction == 1),
        'prediction': 'FAKE' if prediction == 1 else 'REAL',
        'confidence': confidence,
        'fake_prob': prob[1] * 100,
        'real_prob': prob[0] * 100,
        'extracted_text_preview': cleaned[:200] + '...' if len(cleaned) > 200 else cleaned
    }

def predict_from_image(image_path):
    """
    Predict from image input (NEW)
    """
    # Step 1: Extract text from image
    print("🔍 Step 1: Extracting text from image...")
    extracted_text, processed_img = extract_text_from_image(image_path)
    
    if processed_img is None:
        return {
            'error': f'Failed to process image: {extracted_text}',
            'prediction': None,
            'confidence': 0
        }
    
    if len(extracted_text) < 20:
        return {
            'error': 'Not enough text found in image. Try a clearer image.',
            'prediction': None,
            'confidence': 0,
            'extracted_text': extracted_text
        }
    
    print(f"   ✓ Extracted {len(extracted_text)} characters")
    
    # Step 2: Display what was found
    display_image_with_text(image_path, extracted_text)
    
    # Step 3: Predict using extracted text
    print("\n🔍 Step 2: Analyzing extracted text...")
    
    # Try to split into title and body (first line = title, rest = body)
    lines = extracted_text.split('\n')
    title = lines[0] if lines else "Unknown"
    body = ' '.join(lines[1:]) if len(lines) > 1 else extracted_text
    
    # Clean and predict
    cleaned = clean_text(extracted_text)
    vectorized = tfidf.transform([cleaned])
    prediction = model.predict(vectorized)[0]
    prob = model.predict_proba(vectorized)[0]
    confidence = max(prob) * 100
    
    return {
        'input_type': 'IMAGE',
        'is_fake': bool(prediction == 1),
        'prediction': 'FAKE' if prediction == 1 else 'REAL',
        'confidence': confidence,
        'fake_prob': prob[1] * 100,
        'real_prob': prob[0] * 100,
        'extracted_title': title[:100],
        'extracted_body_preview': body[:200] + '...' if len(body) > 200 else body,
        'total_extracted_chars': len(extracted_text)
    }

def display_result(result):
    """
    Pretty print the prediction result
    """
    if 'error' in result and result['error']:
        print(f"\n❌ ERROR: {result['error']}")
        return
    
    print("\n" + "=" * 60)
    print("📊 PREDICTION RESULT")
    print("=" * 60)
    
    # Input type
    input_type = result.get('input_type', 'UNKNOWN')
    print(f"Input Type: {input_type}")
    
    # Main result with colors
    if result['is_fake']:
        print("\n" + "🔴" * 15)
        print("🚨 RESULT: LIKELY FAKE NEWS")
        print("🔴" * 15)
    else:
        print("\n" + "🟢" * 15)
        print("✅ RESULT: LIKELY REAL NEWS")
        print("🟢" * 15)
    
    # Confidence
    print(f"\n📈 Confidence Score: {result['confidence']:.2f}%")
    print(f"   FAKE probability: {result['fake_prob']:.2f}%")
    print(f"   REAL probability: {result['real_prob']:.2f}%")
    
    # Show extracted text preview for images
    if input_type == 'IMAGE':
        print(f"\n📝 Extracted Title: {result.get('extracted_title', 'N/A')}")
        print(f"   Total chars analyzed: {result.get('total_extracted_chars', 0)}")
    
    print("=" * 60)

print("✅ Unified prediction functions ready!")
print("   Use: predict_from_text(title, text) or predict_from_image(image_path)")
```

```python
"""
CELL 18: INTERACTIVE MENU - CHOOSE INPUT TYPE
----------------------------------------------
User selects: Text or Image input
"""

import os

def check_image_path(path):
    """Validate image path"""
    if not path:
        return False, "No path provided"
    if not os.path.exists(path):
        return False, f"File not found: {path}"
    
    valid_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp']
    ext = os.path.splitext(path.lower())[1]
    if ext not in valid_extensions:
        return False, f"Invalid format. Use: {', '.join(valid_extensions)}"
    
    return True, "Valid"

print("=" * 60)
print("🕵️  FAKE NEWS DETECTOR - MULTI-MODE")
print("=" * 60)
print("Choose input type:\n")
print("1️⃣  TEXT - Enter news title and article manually")
print("2️⃣  IMAGE - Upload screenshot/image of news article")
print("q️⃣  QUIT - Exit program\n")

choice = input("Enter choice (1/2/q): ").strip().lower()

if choice == '1':
    # TEXT MODE
    print("\n" + "-" * 60)
    print("📝 TEXT INPUT MODE")
    print("-" * 60)
    
    title = input("\n📰 Enter NEWS TITLE: ")
    print()
    text = input("📝 Enter NEWS TEXT (article content):\n")
    
    if title.strip() and text.strip():
        result = predict_from_text(title, text)
        display_result(result)
    else:
        print("❌ Please provide both title and text!")

elif choice == '2':
    # IMAGE MODE
    print("\n" + "-" * 60)
    print("🖼️  IMAGE INPUT MODE")
    print("-" * 60)
    print("Supported formats: JPG, PNG, BMP, TIFF, WEBP")
    print("Tip: Use clear, high-resolution images for best results\n")
    
    image_path = input("📂 Enter full image path: ").strip()
    
    # Remove quotes if user copied path with quotes
    image_path = image_path.strip('"').strip("'")
    
    # Validate
    is_valid, msg = check_image_path(image_path)
    
    if is_valid:
        print(f"\n✓ Image found: {os.path.basename(image_path)}")
        result = predict_from_image(image_path)
        display_result(result)
    else:
        print(f"❌ {msg}")

elif choice == 'q':
    print("\n👋 Goodbye!")

else:
    print("\n❌ Invalid choice. Please run cell again and select 1, 2, or q.")
```

```python
"""
CELL 19: BATCH IMAGE PROCESSING (OPTIONAL)
-------------------------------------------
Process multiple images at once from a folder
"""

def process_image_folder(folder_path):
    """
    Process all images in a folder
    """
    valid_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp']
    
    # Find all images
    image_files = []
    for f in os.listdir(folder_path):
        ext = os.path.splitext(f.lower())[1]
        if ext in valid_extensions:
            image_files.append(os.path.join(folder_path, f))
    
    if not image_files:
        print(f"No images found in: {folder_path}")
        return
    
    print(f"Found {len(image_files)} images")
    print("=" * 60)
    
    results = []
    
    for i, img_path in enumerate(image_files, 1):
        print(f"\n[{i}/{len(image_files)}] Processing: {os.path.basename(img_path)}")
        print("-" * 40)
        
        result = predict_from_image(img_path)
        display_result(result)
        results.append({
            'file': os.path.basename(img_path),
            'result': result
        })
        
        print("\n" + "=" * 60)
    
    # Summary
    print("\n📊 BATCH PROCESSING SUMMARY")
    print("=" * 60)
    fake_count = sum(1 for r in results if r['result'].get('is_fake', False))
    real_count = len(results) - fake_count
    error_count = sum(1 for r in results if 'error' in r['result'] and r['result']['error'])
    
    print(f"Total images: {len(results)}")
    print(f"   FAKE detected: {fake_count}")
    print(f"   REAL detected: {real_count}")
    print(f"   Errors: {error_count}")

# Uncomment to use:
# folder = input("Enter folder path with images: ").strip()
# process_image_folder(folder)

print("✅ Batch processing function ready!")
print("   Uncomment the last lines to use, or call: process_image_folder('/path/to/folder')")
```
