```python
# Cell 2: Imports
import pandas as pd
import numpy as np
import re
import string
import warnings
import os
warnings.filterwarnings('ignore')

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.pipeline import Pipeline
from scipy.sparse import hstack

import matplotlib.pyplot as plt
import seaborn as sns

try:
    import nltk
    nltk.download('stopwords', quiet=True)
    nltk.download('punkt', quiet=True)
    from nltk.corpus import stopwords
    STOP_WORDS = set(stopwords.words('english'))
    print('✅ Using NLTK stopwords')
except ImportError:
    # Fallback stopwords if NLTK not available
    STOP_WORDS = set("""a an the and or but in on at to for of with is are was were be
    been being have has had do does did will would could should may might shall can
    i me my we our you your he him his she her it its they them their what which
    who this that these those am if""".split())
    print('✅ Using built-in stopwords (NLTK not found)')

print('✅ All imports done!')
```

```python
# Cell 3: Load Dataset
DATASET_PATH = 'fake_or_real_news.csv'

df = pd.read_csv(DATASET_PATH)
print(f'Loaded {len(df)} articles')
print(f'Columns: {list(df.columns)}')
print(f'\nLabel distribution:')
print(df['label'].value_counts())
```

```python
# Cell 4: Fix the Dataset — Strip Template Artifacts
#
# WHY THIS IS THE MOST IMPORTANT STEP:
# Your FAKE articles were generated by taking real news and adding:
#   - Title prefix: 'URGENT: Government hiding evidence that [real title]'
#   - Text prefix: 'Unverified sources claim that...'
#   - Text suffix: '...This has been widely disputed... Experts warn...'
#
# If we DON'T strip these, the model ONLY learns to detect these phrases.
# Any real-world input without these exact phrases = predicted REAL.
# That's why your CNIC scan and other inputs were always classified as REAL.

FAKE_TITLE_PREFIXES = re.compile(
    r'^(URGENT:\s*Government hiding evidence that'
    r'|SHOCKING:\s*What they don\'t want you to know about'
    r'|WARNING:\s*Experts are lying about'
    r'|EXPOSED:\s*Anonymous insider confirms'
    r'|REVEALED:\s*Hidden camera footage shows'
    r'|VIRAL:\s*Everyone is sharing this!'
    r'|BREAKING:\s*Secret sources reveal'
    r'|MUST READ:\s*The truth about'
    r'|LEAKED:\s*Classified documents show'
    r'|BOMBSHELL:\s*|EXCLUSIVE:\s*)\s*',
    re.IGNORECASE
)

FAKE_TEXT_BOILERPLATE = re.compile(
    r'Unverified sources claim that\s*'
    r'|This has been widely disputed by independent fact-checkers\.'
    r'\s*Social media users are sharing this without confirmation\.'
    r'\s*Experts warn this may be misleading information\.',
    re.IGNORECASE
)

def strip_template_artifacts(title, text):
    """Remove synthetic template wrapping from fake articles."""
    title = FAKE_TITLE_PREFIXES.sub('', str(title)).strip()
    text  = FAKE_TEXT_BOILERPLATE.sub('', str(text)).strip()
    return title, text

df['title_clean'], df['text_clean'] = zip(*df.apply(
    lambda r: strip_template_artifacts(r['title'], r['text']), axis=1
))

# Combine: weight title slightly more (appears twice) since it carries more signal
df['combined'] = df['title_clean'] + ' ' + df['title_clean'] + ' ' + df['text_clean']

print('✅ Templates stripped!')
print(f'Example FAKE title before: {df[df["label"]=="FAKE"]["title"].iloc[0][:80]}')
print(f'Example FAKE title after:  {df[df["label"]=="FAKE"]["title_clean"].iloc[0][:80]}')
print(f'\nExample FAKE text before: {df[df["label"]=="FAKE"]["text"].iloc[0][:120]}')
print(f'Example FAKE text after:  {df[df["label"]=="FAKE"]["text_clean"].iloc[0][:120]}')
```

```python
# Cell 5: Sensationalism Feature Engineering
#
# Since our stripped FAKE text is nearly identical to REAL text
# (because FAKE was generated FROM real articles), we add hand-crafted
# linguistic features that distinguish real fake news from real news.

SENSATIONAL_WORDS = [
    # Urgency / alarm words common in fake news
    'shocking', 'breaking', 'urgent', 'exposed', 'revealed', 'leaked',
    'bombshell', 'exclusive', 'viral', 'must read', 'hidden', 'secret',
    'banned', 'censored', 'coverup', 'cover up', 'conspiracy',
    # Unverified attribution
    'anonymous', 'unverified', 'sources say', 'rumor', 'allegedly',
    'reportedly', 'claim', 'insider', 'classified', 'whistleblower',
    # Emotional manipulation
    'outrage', 'outraged', 'furious', 'disgusting', 'horrifying',
    'terrifying', 'devastating', 'unbelievable', 'incredible',
]

CREDIBILITY_PHRASES = [
    # Phrases common in credible reporting
    'according to', 'officials said', 'confirmed by', 'released data',
    'study found', 'report shows', 'government announced', 'press release',
    'ministry of', 'official statement', 'spokesperson said',
]

def extract_features(title, text):
    """
    Extract numeric features beyond TF-IDF.
    Returns a dict of feature values.
    """
    combined = (str(title) + ' ' + str(text)).lower()
    title_lower = str(title).lower()
    
    feats = {}
    
    # Sensationalism score — fake news often uses alarm words
    feats['sensational_count'] = sum(1 for w in SENSATIONAL_WORDS if w in combined)
    
    # Credibility score — reliable news has sourced attribution
    feats['credibility_count'] = sum(1 for p in CREDIBILITY_PHRASES if p in combined)
    
    # Exclamation marks — rare in real journalism
    feats['exclamations'] = combined.count('!')
    
    # All-caps words — sensationalism signal
    feats['caps_words'] = len(re.findall(r'\b[A-Z]{3,}\b', str(title)))
    
    # Question marks in title — clickbait pattern
    feats['title_questions'] = title_lower.count('?')
    
    # Numbers in text — real news often has specific data
    feats['numbers_in_text'] = len(re.findall(r'\d+', str(text)))
    
    # Text length — very short text is suspect
    feats['text_length'] = len(str(text).split())
    
    return feats

feature_df = df.apply(
    lambda r: extract_features(r['title_clean'], r['text_clean']), axis=1, result_type='expand'
)
print('✅ Linguistic features extracted:')
print(feature_df.describe())
```

```python
# Cell 6: Text Cleaning

def clean_text(text):
    """Clean and normalize text for TF-IDF."""
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)   # Remove URLs
    text = re.sub(r'\S+@\S+', '', text)                    # Remove emails
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    words = text.split()
    words = [w for w in words if w not in STOP_WORDS and len(w) > 2]
    return ' '.join(words)

df['processed'] = df['combined'].apply(clean_text)
df = df[df['processed'].str.len() > 10].copy().reset_index(drop=True)

# Reindex feature_df to match
feature_df = feature_df.loc[df.index].reset_index(drop=True)
df = df.reset_index(drop=True)

print(f'✅ Text cleaned. Articles remaining: {len(df)}')
print(f'Label distribution: {df["label"].value_counts().to_dict()}')
```

```python
# Cell 7: Train-Test Split
from sklearn.model_selection import train_test_split
from scipy.sparse import hstack, csr_matrix

X_text = df['processed']
y = df['label']

# Split indices so we can split both text and numeric features
idx_train, idx_test = train_test_split(
    df.index, test_size=0.2, random_state=42, stratify=y
)

print(f'Train: {len(idx_train)}, Test: {len(idx_test)}')
print(f'Train — REAL: {(y[idx_train]=="REAL").sum()}, FAKE: {(y[idx_train]=="FAKE").sum()}')
print(f'Test  — REAL: {(y[idx_test]=="REAL").sum()}, FAKE: {(y[idx_test]=="FAKE").sum()}')
```

```python
# Cell 8: Build Feature Matrix (TF-IDF + Linguistic Features)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from scipy.sparse import hstack, csr_matrix

# TF-IDF on cleaned text
tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),      # unigrams + bigrams
    min_df=2,                 # ignore very rare terms
    max_df=0.85,              # ignore near-universal terms
    sublinear_tf=True         # log scaling of TF
)

X_tfidf_train = tfidf.fit_transform(X_text[idx_train])
X_tfidf_test  = tfidf.transform(X_text[idx_test])

# Normalize linguistic features
scaler = StandardScaler()
X_feats_train = csr_matrix(scaler.fit_transform(feature_df.loc[idx_train]))
X_feats_test  = csr_matrix(scaler.transform(feature_df.loc[idx_test]))

# Combine TF-IDF + linguistic features
X_train = hstack([X_tfidf_train, X_feats_train])
X_test  = hstack([X_tfidf_test,  X_feats_test])
y_train = y[idx_train]
y_test  = y[idx_test]

print(f'✅ Feature matrix: {X_train.shape[1]} features total')
print(f'   TF-IDF features: {X_tfidf_train.shape[1]}')
print(f'   Linguistic features: {feature_df.shape[1]}')
```

```python
# Cell 9: Train the Model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = LogisticRegression(
    max_iter=1000,
    random_state=42,
    class_weight='balanced',  # handles FAKE/REAL imbalance
    C=0.5                     # slightly more regularization for small dataset
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f'🎯 Test Accuracy: {accuracy*100:.2f}%')
print()
print('📊 Classification Report:')
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred, labels=['FAKE', 'REAL'])
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['FAKE', 'REAL'], yticklabels=['FAKE', 'REAL'])
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
```

```python
# Cell 10: THE FIXED PREDICTION FUNCTION
#
# Key changes from original:
# 1. Strips template artifacts (the main bug fix)
# 2. Adds linguistic feature extraction
# 3. Returns UNCERTAIN for borderline predictions (50-60% confidence)
# 4. Clear warning when text is too short/irrelevant

def predict_news(title, text, confidence_threshold=0.60):
    """
    Predict whether a news article is FAKE or REAL.
    
    Parameters:
    -----------
    title : str — Headline of the article
    text  : str — Body text of the article
    confidence_threshold : float — Minimum confidence to make a call (default 60%)
    
    Returns:
    --------
    dict with prediction, confidence, probabilities, and warnings
    """
    # --- Step 1: Strip synthetic template artifacts (THE KEY FIX) ---
    title_clean = FAKE_TITLE_PREFIXES.sub('', str(title)).strip()
    text_clean  = FAKE_TEXT_BOILERPLATE.sub('', str(text)).strip()
    
    # --- Step 2: Check if there's enough text to analyze ---
    combined_raw = title_clean + ' ' + text_clean
    word_count = len(combined_raw.split())
    
    warnings_list = []
    if word_count < 10:
        warnings_list.append(
            f'⚠️ Very short input ({word_count} words). '
            'Results may not be reliable. This model is designed for news articles, '
            'not ID cards or very short text.'
        )
    
    # --- Step 3: Extract linguistic features ---
    feats = extract_features(title_clean, text_clean)
    feats_scaled = scaler.transform([list(feats.values())])
    
    # --- Step 4: TF-IDF + predict ---
    cleaned = clean_text(combined_raw)
    tfidf_vec = tfidf.transform([cleaned])
    X_input = hstack([tfidf_vec, csr_matrix(feats_scaled)])
    
    raw_pred = model.predict(X_input)[0]
    probs = model.predict_proba(X_input)[0]
    prob_dict = {cls: round(p * 100, 1) for cls, p in zip(model.classes_, probs)}
    
    confidence = max(probs)
    
    # --- Step 5: Apply confidence threshold ---
    if confidence < confidence_threshold:
        final_pred = 'UNCERTAIN'
        warnings_list.append(
            f'⚠️ Low confidence ({confidence*100:.1f}%). '
            'The model is not sure — treat result with caution.'
        )
    else:
        final_pred = raw_pred
    
    return {
        'prediction': final_pred,
        'is_fake': final_pred == 'FAKE',
        'confidence': round(confidence * 100, 1),
        'probabilities': prob_dict,
        'linguistic_signals': feats,
        'warnings': warnings_list
    }


def display_result(result, title=''):
    """Pretty-print prediction result."""
    print('\n' + '='*60)
    if title:
        print(f'📰 {title[:55]}')
        print('-'*60)
    
    pred = result['prediction']
    if pred == 'FAKE':
        print('🔴 RESULT: FAKE NEWS')
    elif pred == 'REAL':
        print('🟢 RESULT: REAL NEWS')
    else:
        print('🟡 RESULT: UNCERTAIN')
    
    print(f'   Confidence  : {result["confidence"]}%')
    print(f'   FAKE prob   : {result["probabilities"].get("FAKE", 0)}%')
    print(f'   REAL prob   : {result["probabilities"].get("REAL", 0)}%')
    
    sigs = result['linguistic_signals']
    print(f'   Sensational words : {sigs["sensational_count"]}')
    print(f'   Credibility phrases: {sigs["credibility_count"]}')
    print(f'   Exclamation marks : {sigs["exclamations"]}')
    
    for w in result['warnings']:
        print(f'   {w}')
    print('='*60)

print('✅ Prediction function ready!')
```

```python
# Cell 11: Test with Examples

test_cases = [
    {
        'title': 'Iran Israel war',
        'text': 'Iran has been killed nytan yaho the president of israel. i was listening this on social media'
    },
    {
        'title': 'Donald Trump martyred',
        'text': 'Iran killed donald trump. Shocking news nobody is talking about this.'
    },
    {
        'title': 'Govt hikes petrol price',
        'text': 'The government has increased petrol prices by Rs10 per litre according to official notification.'
    },
    {
        'title': 'Federal Reserve Announces Interest Rate Decision',
        'text': 'The Federal Reserve announced today that interest rates will remain unchanged. '
                'Chairman Powell cited moderate inflation and stable employment figures as the basis for the decision.'
    },
    {
        'title': 'SHOCKING: Scientists Discover Miracle Cure Doctors HIDE from You!!!',
        'text': 'Big pharma does NOT want you to know this secret cure. Anonymous insiders leaked '
                'classified documents showing this banned remedy cures all diseases! Share before they delete this!'
    },
    {
        'title': 'CNIC 12345-1234567-1',
        'text': 'Name: Ali Khan. Date of Birth: 15-01-1990. Address: House 5 Street 3 Lahore Punjab Pakistan.'
    },
]

for case in test_cases:
    result = predict_news(case['title'], case['text'])
    display_result(result, title=case['title'])
```

```python
# Cell 12: Interactive Mode

print('='*60)
print('🕵️  FAKE NEWS DETECTOR — FIXED INTERACTIVE MODE')
print('='*60)
print("Type 'quit' in the title field to exit\n")

while True:
    user_title = input('\n📰 Enter NEWS TITLE (or "quit" to exit): ').strip()
    
    if user_title.lower() in ('quit', 'exit', 'q'):
        print('\n👋 Goodbye!')
        break
    
    user_text = input('📝 Enter NEWS TEXT: ').strip()
    
    if not user_title and not user_text:
        print('⚠️  Please enter at least a title or text.')
        continue
    
    result = predict_news(user_title, user_text)
    display_result(result, title=user_title)
```

```python
# Cell 13: Save the Fixed Model
import joblib

os.makedirs('saved_models', exist_ok=True)

model_package = {
    'model': model,
    'tfidf': tfidf,
    'scaler': scaler,
    'stop_words': list(STOP_WORDS),
    'accuracy': accuracy,
    'feature_names': list(feature_df.columns),
    'version': 'fixed_v2',
}

joblib.dump(model_package, 'saved_models/fake_news_model_fixed.pkl')
print('✅ Fixed model saved to: saved_models/fake_news_model_fixed.pkl')
print(f'   Accuracy: {accuracy*100:.2f}%')
```
